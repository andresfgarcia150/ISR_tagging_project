```python
%pylab inline
caseForFile = 'sTopsWithISR'
caseStudied = '\n$sTops\ with\ ISR\ Jet$'
```

```python
# Creating the arrays Pcut and PT, and the matrix tags
Pcutarray = []
PTarray = []
tags = []
missTags = []
rejected = []
f = open('../resultsTagging-3e-7/tags_sTopsWISR.txt','r')
fm = open('../resultsTagging-3e-7/misstags_sTopsWISR.txt','r')
fr = open('../resultsTagging-3e-7/rejected_sTopsWISR.txt','r')
f.readline()
fm.readline()
fr.readline()
fm.readline()
fr.readline()
PTtext = f.readline().split('     	')

for i in range(len(PTtext))[1:len(PTtext)+1]:
    if(i == len(PTtext)-1):
        temp = PTtext[i].split('	 ')
        PTarray.append(float(temp[0]))
    else:
        PTarray.append(float(PTtext[i]))
    
for line in f:
    l = line.split('     ')
    tempTags = []
    for i in range(len(l)):
        if (i == 0):
            Pcutarray.append(float(l[i]))
        else:
            temp = l[i].replace('\t','')
            temp = temp.replace('%','')
            tempTags.append(float(temp))
    tags.append(tempTags)
    
for line in fm:
    l = line.split('     ')
    tempMissTags = []
    for i in range(len(l)):
        if (i != 0):
            temp = l[i].replace('\t','')
            temp = temp.replace('%','')
            tempMissTags.append(float(temp))
    missTags.append(tempMissTags)

for line in fr:
    l = line.split('     ')
    tempRejected = []
    for i in range(len(l)):
        if (i != 0):
            temp = l[i].replace('\t','')
            temp = temp.replace('%','')
            tempRejected.append(float(temp))
    rejected.append(tempRejected)
    
f.close()

tags = array(tags)
missTags = array(missTags)
rejected = array(rejected)
print Pcutarray
print PTarray
print tags[0]
print missTags[0]
print rejected[0]
filas = shape(tags)[0]
cols = shape(tags)[1]
print cols, filas
```

```python
hold
figure(figsize=(6,5))
plot(Pcutarray,tags[:,cols-1])
plot(Pcutarray,missTags[:,cols-1],'r')
plot(Pcutarray,rejected[:,cols-1],'g')
axis([1.0,3.0,0,100])
xlabel('$k\ (P_{cut}=k/N_j)$',fontsize=16)
ylabel('$[\%]$',fontsize=16)
string = '$Overall\ percentages$' + caseStudied
title(string,fontsize=16);
legend(["Tags", "MissTags", "Rejected"],loc=2)
nameFig = './'+ caseForFile + '/OverallPercentages_' + caseForFile + '.jpg'
savefig(nameFig,dpi=200)
```

```python
print 'PT Tagsmin Tagsmax Misstagsmin Misstagsmax'
vect = [6,10,12,15,18]
for i in vect:
    currentCol = i
    figure(figsize=(6,5))
    hold
    plot(Pcutarray,tags[:,currentCol])
    plot(Pcutarray,missTags[:,currentCol],'r')
    plot(Pcutarray,rejected[:,currentCol],'g')
    axis([1.0,3.0,0,100])
    xlabel('$k\ (P_{cut}=k/N_j)$',fontsize=16)
    ylabel('$[\%]$',fontsize=16)
    string = '$Percentages\ PT_{cut}\ =\ $' + str(PTarray[currentCol]) + '$GeV$' + caseStudied
    title(string,fontsize=16);
    legend(["Tags", "MissTags", "Rejected"],loc=2)
    nameFig = './'+ caseForFile + '/Percentages_PT_' + str(PTarray[currentCol]) + caseForFile + '.jpg'
    savefig(nameFig,dpi=200)
    print ("%.1f &" % PTarray[currentCol] + " %.2f &" % amin(tags[:,currentCol]) + " %.2f &" % amax(tags[:,currentCol]) + " %.2f &" % amin(missTags[:,currentCol]) + " %.2f &" % amax(missTags[:,currentCol]) )
```

```python
lvls = linspace(0,amax(tags),50)
figure(figsize=(6.0,5.2))
contourf(PTarray,Pcutarray,tags[:,0:cols-1],levels = lvls)
xlabel('$PT_{cut}\ [GeV]$',fontsize=16)
ylabel('$k\ (P_{cut}=k/N_j)$',fontsize=16)
string = '$Percentage\ of\ Correct\ Tags$' + caseStudied
title(string,fontsize=16);
colorbar()
nameFig = './'+ caseForFile + '/Correct_Tags_' + caseForFile + '.jpg'
savefig(nameFig,dpi=200)

lvls = linspace(0,amax(missTags),50)
figure(figsize=(6.0,5.2))
contourf(PTarray,Pcutarray,missTags[:,0:cols-1],levels = lvls)
xlabel('$PT_{cut}\ [GeV]$',fontsize=16)
ylabel('$k\ (P_{cut}=k/N_j)$',fontsize=16)
string = '$Percentage\ of\ Incorrect\ Tags$' + caseStudied
title(string,fontsize=16);
colorbar()
nameFig = './'+ caseForFile + '/Incorrect_Tags' + caseForFile + '.jpg'
savefig(nameFig,dpi=200)

lvls = linspace(0,amax(rejected),50)
figure(figsize=(6.0,5.2))
contourf(PTarray,Pcutarray,rejected[:,0:cols-1],levels = lvls)
xlabel('$PT_{cut}\ [GeV]$',fontsize=16)
ylabel('$k\ (P_{cut}=k/N_j)$',fontsize=16)
string = '$Percentage\ of\ Rejected\ Events$' + caseStudied
title(string,fontsize=16);
colorbar()
nameFig = './'+ caseForFile + '/Rejected_Tags' + caseForFile + '.jpg'
savefig(nameFig,dpi=200)
```
