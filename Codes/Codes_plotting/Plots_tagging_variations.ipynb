# Plots of the overall results of the tagging algorithm

```python
%pylab inline
plt.ioff() # Does not show the pictures automatically
# Obs: Write show() if you want to plot on the screen all the figures that have not been closed (using close(figure))

Tops_o_sTops = 's' #  '_' for Tops and 's' for sTops
ISR_or_not = 'SI' # 'WI' With ISR, 'SI' without ISR, 'bb' for btags
matching = '_Matching' # '_Matching' if a matching procedure has been done, '' otherwise
analysis = 'sTops\ With\ ISR$'

if (Tops_o_sTops == '_'):
    analysis = '\ '
else:
    analysis = 's'
    
if (ISR_or_not == 'WI'):
    analysis = analysis + 'Tops\ with\ ISR$'
elif (ISR_or_not == 'SI'):
    analysis = analysis + 'Tops\ without\ ISR$'
elif (ISR_or_not == 'bb'):
    analysis = analysis + 'Tops\ with\ b\ jets\ and\ ISR\ jet$'

local_folder = Tops_o_sTops + 'Tops_result_' + ISR_or_not + matching + '/'

# The name of the studied variables
variables = []
variables.append('$PT$')
variables.append('$|\eta|$')
variables.append('$\Delta \Phi_{MET-Jet}$')
variables.append('$PT_{ratio}$')
variables.append('$\Delta \eta_{aver}$')
variables.append('$\Delta \Phi_{MET-Jet-Others}$')
variables.append('$\Delta PT_{leading}$')
variables.append('$\Delta \eta_{leading}$')
```

### Reading the input files

```python
# Creating the arrays k_values, tags, missTags, rejected
k_values = []
tags = []
missTags = []
rejected = []

# Reading the tagging results
max_var = 7 # Last variable
for i in range(0,max_var):
    for j in range(i+1,max_var):
        for k in range(j+1,max_var):
            comb = '_' + str(i) + '_' + str(j) + '_' + str(k)
            file_Name = '../../Results/resultsTagging/' + local_folder + Tops_o_sTops + 'Tops_' + ISR_or_not + '_Overall' + comb  + '.txt'

            fl = open(file_Name)
            
            # Reading comments
            fl.readline()
            fl.readline()
            fl.readline()
            
            tempTags = []
            tempMiss = []
            tempReje = []
            
            for line in fl:
                l = line.split('	')
                tempTags.append(float(l[1]))
                tempMiss.append(float(l[2]))
                tempReje.append(float(l[3]))
                if ((i == 0) & (j == 1) & (k == 2)):
                    k_values.append(float(l[0]))
                
            tags.append(tempTags)
            missTags.append(tempMiss)
            rejected.append(tempReje)

tags = array(tags)
missTags = array(missTags)
rejected = array(rejected)
print (k_values)
print (tags[0])
```

### Plotting the tagging results

```python
itera = 0
for i in range(0,max_var):
    for j in range(i+1,max_var):
        for k in range(j+1,max_var):
            comb = '_' + str(i) + '_' + str(j) + '_' + str(k)

            hold
            fig_comp = figure(figsize=(6,5))
            if ((ISR_or_not == 'WI') or (ISR_or_not == 'bb')):
                plot(k_values,tags[itera,:])
            plot(k_values,missTags[itera,:],'r')
            plot(k_values,rejected[itera,:],'g')
            axis([1.0,3.0,0,1])
            xlabel('$k\ (P_{cut}=k/N_j)$',fontsize=16)
            ylabel('$[\%]$',fontsize=16)
            string = '$Overall\ percentages\ Case:\ ' + analysis + '\n$Variables:\ $' + variables[i] + ' - ' + variables[j] + ' - ' + variables[k]
            title(string,fontsize=16);
            if ((ISR_or_not == 'WI') or (ISR_or_not == 'bb')):
                legend(["Tags", "MissTags", "No Tags"],loc=2)
            elif (ISR_or_not == 'SI'):
                legend(["MissTags", "No Tags"],loc=7)
            nameFig = '../../Results/plots_tagging_variables_comparison/' + local_folder + 'OverallPercentages' + Tops_o_sTops + 'Tops_' + ISR_or_not + comb + '.jpg'
            savefig(nameFig,dpi=200)
            itera += 1
            
            if (itera == 1):
                show()
            
            close(fig_comp)
```

### Finding the optimum set of variables

```python
diff_Tag_Miss = (tags - missTags)/missTags

print (diff_Tag_Miss[0,0])
```

```python
leyend_seq = []
itera = 0

if ((ISR_or_not == 'WI') or (ISR_or_not == 'bb')):
    hold
    fig_diff = figure(figsize=(15,15))
    for i in range(0,max_var):
        for j in range(i+1,max_var):
            for k in range(j+1,max_var):
                plot(k_values,diff_Tag_Miss[itera,:])
                leyend_seq.append(variables[i] + ' - ' + variables[j] + ' - ' + variables[k])
                itera += 1
                
    axis([0.2,3.0,-1.0,15.0])
    xlabel('$k\ (P_{cut}=k/N_j)$',fontsize=16)
    ylabel('${(Tags-Misstags)}/{Misstags}$',fontsize=16)
    string = '${(Tags-Misstags)}/{Misstags}$\n' + '$ Case:\ ' + analysis
    title(string,fontsize=16);
    legend(leyend_seq,loc=2)
    nameFig = '../../Results/plots_tagging_variables_comparison/' + local_folder + Tops_o_sTops + 'Tops_' + ISR_or_not + '_Comparison_between_different_set_of_variables.jpg'
    savefig(nameFig,dpi=200)
    show()
    close(fig_diff)
    
elif  (ISR_or_not == 'SI'):
    hold
    fig_diff = figure(figsize=(15,15))
    for i in range(0,max_var):
        for j in range(i+1,max_var):
            for k in range(j+1,max_var):
                plot(k_values,rejected[itera,:])
                leyend_seq.append(variables[i] + ' - ' + variables[j] + ' - ' + variables[k])
                itera += 1
                
    axis([0.2,3.0,0.0,1.0])
    xlabel('$k\ (P_{cut}=k/N_j)$',fontsize=16)
    ylabel('$No\ Tags$',fontsize=16)
    string = '$No\ Tags$\n' + '$ Case:\ ' + analysis
    title(string,fontsize=16);
    legend(leyend_seq,loc=2)
    nameFig = '../../Results/plots_tagging_variables_comparison/' + local_folder + Tops_o_sTops + 'Tops_' + ISR_or_not + '_Comparison_between_different_set_of_variables.jpg'
    savefig(nameFig,dpi=200)
    show()
    close(fig_diff)

```

##### Combination of variables that distinguish, distinguish partially and do not distinguish

```python
# Looking at the chart bellow, a set of variables with diff_Tag_Miss greater than 0.2 @ k=1.2 distinguish efficiently 
# Likewise, if diff_Tag_Miss is bellow than -0.3 @ k=1.2, the set does not distinguish
# Finally, if diff_Tag_Miss is between -0.3 and 0.2 @ k=1.2, the set distinguis partially

# On the other hand, in those events without ISR jets, the sets of variables seem to be grouped in 3 groups according 
# to the overall performance. The best of them presents no tags > 0.95 @ k = 2.0. The worst, No tags < 0.92 @ k = 2.0 
# for Tops and 0.85 for sTops

file_Name = '../../Results/plots_tagging_variables_comparison/' + local_folder + Tops_o_sTops + 'Tops_' + ISR_or_not + '_Variables_performance.txt'
fl = open(file_Name,'w')

if (ISR_or_not == 'WI'):
        
    val = 0 # Evaluation value
    min_val = 0
    max_val = 0
    if (Tops_o_sTops == '_'):
        max_val = 0.2
        min_val = -0.3
    else:
        max_val = 0.2
        min_val = -0.2

    disting = []
    part_dist = []
    not_dist = []
    
    itera = 0
    for i in range(0,max_var):
        for j in range(i+1,max_var):
            for k in range(j+1,max_var):
                comb = str(i) + '-' + str(j) + '-' + str(k) + '\t'
                if (diff_Tag_Miss[itera,val] > max_val):
                    disting.append(comb + '-' + variables[i] + ' - ' + variables[j] + ' - ' + variables[k])
                elif (diff_Tag_Miss[itera,val] < min_val):
                    not_dist.append(comb + '-' + variables[i] + ' - ' + variables[j] + ' - ' + variables[k])
                else:
                    part_dist.append(comb + '-' + variables[i] + ' - ' + variables[j] + ' - ' + variables[k])      
                itera += 1
    
    chain = '\nDistinguish:\n'
    print chain
    fl.write(chain)
    for i in range(0,len(disting)):
        print disting[i]
        fl.write(disting[i]+'\n')
        
    chain = '\n\nDistinguish partially:\n'
    print chain
    fl.write(chain)
    for i in range(0,len(part_dist)):
        print part_dist[i]
        fl.write(part_dist[i]+'\n')
        
    chain = '\n\nDo not distinguish:\n'
    print chain
    fl.write(chain)
    for i in range(0,len(not_dist)):
        print not_dist[i]
        fl.write(not_dist[i]+'\n')
        
elif  (ISR_or_not == 'SI'):
    min_val = 0
    max_val = 0
    if (Tops_o_sTops == '_'):
        max_val = 0.96
        min_val = 0.92
    else:
        max_val = 0.95
        min_val = 0.85
           
    best = []
    second = []
    worst = []
    
    itera = 0
    for i in range(0,max_var):
        for j in range(i+1,max_var):
            for k in range(j+1,max_var):
                comb = str(i) + '-' + str(j) + '-' + str(k) + '\t'
                val = int((2.0-1.2)/(k_values[2]-k_values[1]))
                if (rejected[itera,val] > max_val):
                    best.append(comb + '-' + variables[i] + ' - ' + variables[j] + ' - ' + variables[k])
                elif (rejected[itera,val] < min_val):
                    worst.append(comb + '-' + variables[i] + ' - ' + variables[j] + ' - ' + variables[k])
                else:
                    second.append(comb + '-' + variables[i] + ' - ' + variables[j] + ' - ' + variables[k])      
                itera += 1
    
    chain = '\nBest performance:\n'
    print chain
    fl.write(chain)
    for i in range(0,len(best)):
        print best[i]
        fl.write(best[i]+'\n')
        
    chain = '\n\nSecond best performance:\n'
    print chain
    fl.write(chain)
    for i in range(0,len(second)):
        print second[i]
        fl.write(second[i]+'\n')
        
    chain = '\n\nWorst performance:\n'
    print chain
    fl.write(chain)
    for i in range(0,len(worst)):
        print worst[i]
        fl.write(worst[i]+'\n')
        
elif (ISR_or_not == 'bb'):
        
    val = int((1.5-1.2)/(k_values[1]-k_values[0])) # Evaluation value
    min_val = 0.25
    max_val = 0.75       

    disting = []
    part_dist = []
    not_dist = []
    
    itera = 0
    for i in range(0,max_var):
        for j in range(i+1,max_var):
            for k in range(j+1,max_var):
                comb = str(i) + '-' + str(j) + '-' + str(k) + '\t'
                if (diff_Tag_Miss[itera,val] > max_val):
                    disting.append(comb + '-' + variables[i] + ' - ' + variables[j] + ' - ' + variables[k])
                elif (diff_Tag_Miss[itera,val] < min_val):
                    not_dist.append(comb + '-' + variables[i] + ' - ' + variables[j] + ' - ' + variables[k])
                else:
                    part_dist.append(comb + '-' + variables[i] + ' - ' + variables[j] + ' - ' + variables[k])      
                itera += 1
    
    chain = '\nDistinguish:\n'
    print chain
    fl.write(chain)
    for i in range(0,len(disting)):
        print disting[i]
        fl.write(disting[i]+'\n')
        
    chain = '\n\nDistinguish partially:\n'
    print chain
    fl.write(chain)
    for i in range(0,len(part_dist)):
        print part_dist[i]
        fl.write(part_dist[i]+'\n')
        
    chain = '\n\nDo not distinguish:\n'
    print chain
    fl.write(chain)
    for i in range(0,len(not_dist)):
        print not_dist[i]
        fl.write(not_dist[i]+'\n')

fl.close()
```
